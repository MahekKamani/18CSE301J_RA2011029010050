```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(weather['Humidity9am'], weather['MaxTemp'])
```

```python
sns.distplot(weather['Humidity9am'])
```

```python
sns.distplot(weather['Humidity9am'], kde=False, rug=True)
```

```python
sns.jointplot(weather['Humidity3pm'], weather['MaxTemp'])
```

```python
sns.jointplot(weather['Humidity3pm'], weather['MaxTemp'], kind='hex')
```

```python
sns.jointplot(weather['Humidity3pm'], weather['MaxTemp'], kind='kde')
```

```python
sns.stripplot(weather['Temp9am'],weather['RainTomorrow'])
```

```python
sns.swarmplot(weather['Humidity3pm'], weather['MaxTemp'])
```
